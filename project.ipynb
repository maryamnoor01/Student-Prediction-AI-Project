<div class="alert alert-block alert-info">
    5. <b>Train - Test Splitting</b>
</div>

```python
X = df.drop('Gender', axis=1)   # Dependant columns / Input columns
y = df['Gender']                # Resultant columns / Output column
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False) # 80:20, 70:30, 60:40
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

<div class="alert alert-block alert-info">
    6. <b>Applying Classifier / Model Training</b>
</div>

```python
from sklearn.svm import SVC # Support Vector Classifier

model_svc = SVC()
model_svc.fit(train_X, train_y)

print(model_svc)
```

```python
model_pred_svc = model_svc.predict(test_X)
model_pred_svc
```

```python
from sklearn.metrics import accuracy_score
model_acc_svc = accuracy_score(test_y, model_pred_svc)
print(round(model_acc_svc,3)*100, "%")
```
